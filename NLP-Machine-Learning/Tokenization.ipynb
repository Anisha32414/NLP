```python
corpus="""Hello Welcome, to this NLP tutorial. 
Please do practice all the methods used to preprocess data in NPL.
It's a very good practice to learn NLP.
This repository consist of various notebooks that will help you. 
I hope you like it and find helpful."""
```

```python
print(corpus)
```

```python
import nltk

nltk.download('punkt_tab')
nltk.download('punkt')
```

```python
## Tokenization: paragraph --> sentences

''' sent_tokenize() function is used to split the text into sentences. 
It takes a string as input and returns a list of sentences.'''

from nltk.tokenize import sent_tokenize
doc=sent_tokenize(corpus)
print(type(doc))
print(doc)
```

```python
for sentences in doc:
    print(sentences)
```

```python
## Tokenization: sentences --> words

''' word_tokenize() function is used to split the text into words.'''

from nltk.tokenize import word_tokenize
words=word_tokenize(corpus)
print(words)
```

```python
for sentence in doc:
    line_wise_words=word_tokenize(sentence)
    print(line_wise_words)
```

```python
""" In this type ''s' are considered different words as ["'" , "s"]. """

from nltk.tokenize import wordpunct_tokenize
words_with_punctuation=wordpunct_tokenize(corpus)
print(words_with_punctuation)
```

```python
"""In this type middle '.' are not considered as a punctuation. 
It is considered as a part of the word. Only the Last '.' is considered as a punctuation."""

from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
