### Stemming

```python
words=["eating","eat","eaten", "sleeping", "slept","sleep", "running","run", "jumping","jumped","jumps", "reading","read", "writing","wrote", "coding","coded", "playing","play","played"]
```

### PorterStemmer

```python
from nltk.stem import PorterStemmer
stemming=PorterStemmer()

for word in words:
    root=stemming.stem(word)
    print(word+"--->"+root)
```

```python
## There are some issues with stemming that we can see it changed the root word of some words.

stemming.stem("congratulation")  # it should have been congrat but it is congratul which is a problem
```

### RegexpStemmer Class

```python
''' It will remove the ing, s, e, able from the end of the word if it is present 
and the length of the word is greater than 4. 
It will not remove the ing, s, e, able if the length of the word is less than 4 or from the front.'''

from nltk.stem import RegexpStemmer
reg_stem=RegexpStemmer('ing$|s$|e$|able$', min=4)
reg_stem.stem("eating")  # it should have been congrat but it is congratul which is a problem
```

```python
reg_stem.stem("ingeating") # it will only remove ing from end 
```

```python
## for removing the ing, s, e, able from the front we can use the following code

from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('^ing|^s|^e|^able', min=4)
reg_stemmer.stem("ingeating") # it will only remove ing from front
```

```python
## fro removing the ing, s, e, able from both front and end we can use the following code

from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing', min=4)
reg_stemmer.stem("ingeating") # it will remove ing from both front and end  
```

### Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
snowball_stemmer=SnowballStemmer('english')

for word in words:
    root_word=snowball_stemmer.stem(word)
    print(word+"-->"+root_word)
```

```python
## PorterStemmer gives following result

print(stemming.stem("fairly"))
print(stemming.stem("sportingly"))
```

```python
## SnowballStemmer gives following results for the same input words
print(snowball_stemmer.stem("fairly"))
print(snowball_stemmer.stem("sportingly"))
```

```python
## But still SnowballStemmer gives some issues with some words

print(snowball_stemmer.stem("goes"))
print(snowball_stemmer.stem("dancing"))
```
