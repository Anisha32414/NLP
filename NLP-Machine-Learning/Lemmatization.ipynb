### Drawbacks of Stemming is covered in Lemmatization

#### Wordnet Lemmatizer

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
'''
POS:-
Noun -n
verb -v
adjective -a
adverb -r
'''
lemmatizer.lemmatize("doing",pos='v')
```

```python
words=["eating","eat","eaten", "sleeping", "slept","sleep", "running","run", "jumping","jumped","jumps", "reading","read", "writing","wrote", "coding","coded", "playing","play","played"]
```

```python
for word in words:
    root_word=lemmatizer.lemmatize(word,pos='v')
    print(word+"-->"+root_word)
```
