## Stop Words in NLP

```python
paragraph="Today, I stand before you to talk about a force that is quietly but " \
"radically reshaping our world: Artificial Intelligence. " \
"Once relegated to the realm of science fiction, AI is now the invisible engine driving our daily " \
"lives. From the moment we wake up and check our phones with facial recognition, to the routes " \
"we take using maps, and the entertainment we stream, AI is constantly analyzing data, learning " \
"patterns, and making decisions to make our lives easier and more efficient. " \
"Beyond our personal conveniences, AI is tackling some of humanity's biggest challenges. " \
"In healthcare, machine learning algorithms are helping doctors detect diseases like cancer at " \
"much earlier stages. In education, AI is personalizing learning experiences—adapting to a " \
"student's unique pace and needs. It is also transforming industries from agriculture to finance, " \
"making processes smarter and more sustainable."
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer

stemmer=PorterStemmer()
sentences=nltk.sent_tokenize(paragraph)
```

```python
## Apply Stopwords and filter and the apply stemming

for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    filtered_words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(filtered_words)
    ## converting all the list of words into sentences
```

```python
## Paragraph after filtering
print(sentences)
```

```python
## Now doing same process using Snowball Stemmer as we did not get good results
from nltk.stem import SnowballStemmer
snowball_stemmer=SnowballStemmer('english')
sentences=nltk.sent_tokenize(paragraph)
```

```python
## Apply Stopwords and filter and the apply Snowball stemming

for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    filtered_words=[snowball_stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(filtered_words)
```

```python
print(sentences)
```
